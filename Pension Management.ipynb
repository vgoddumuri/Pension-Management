```python
# Missing benificiary, percentage
total_rows = len(df)
missing_count = df['beneficiary_name'].isna().sum()
missing_pct = (missing_count / total_rows) * 100

print(f"Total Rows: {total_rows}")
print(f"Missing beneficiary_name: {missing_count}")
print(f"Missing Percentage: {missing_pct:.2f}%")
```

```python
df['beneficiary_name_missing'] = df['beneficiary_name'].isna().astype(int) # Create missing indicator column
```

```python
# Clean & standardize text
for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.strip()          # remove spaces
    df[col] = df[col].str.replace("  +", " ", regex=True)
```

```python
# Adding Beneficiary Not Registered if Benificiary name null
df['beneficiary_name']=df['beneficiary_name'].fillna('Beneficiary Not Registered')
```

```python
df.head()
```

```python
df['dob'] = pd.to_datetime(df['dob'], errors='coerce')
df['retirement_date'] = pd.to_datetime(df['retirement_date'], errors='coerce')
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce')
```

```python
# calculating age at Retirement

df['age_at_retirement'] = ((df['retirement_date'] - df['dob']).dt.days // 365)
df.head()
```

```python
#Remove invalid ages
df.loc[df['age_at_retirement'] < 0, 'age_at_retirement'] = None
```

```python
df['pension_status'] = 'Not Retired'
df.loc[df['retirement_date'].notna(), 'pension_status'] = 'Pending'
df.loc[df['join_date'].notna(), 'pension_status'] = 'Active'
```

```python
df['data_issue_flag'] = 0
df.loc[df['join_date'] < df['retirement_date'], 'data_issue_flag'] = 1
```
